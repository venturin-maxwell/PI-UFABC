O programa deve ler 5 notas: três testes (um a um) e duas provas (uma a uma).

Para chegar ao resultado, considere o peso de cada avaliação no cálculo da Nota Final (NF):

Como calcular a média aritmética T entre os três testes para que ela represente 20% da nota final?
Sabendo que cada prova vale 40%, qual a expressão matemática que une T, P1 e P2?
Ao classificar o conceito (A, B, C, D ou F), como você estruturaria os testes para garantir que cada nota caia na faixa correta sem sobreposição?

```python
T1 = float(input())
T2 = float(input())
T3 = float(input())
P1 = float(input())
P2 = float(input())

Provas = (P1+P2)*.4
T_media = ((T1+T2+T3)/(3))*.2

conceito = T_media + Provas

if conceito >= 9 <= 10:
    print('{:.2f}'.format(conceito))
    print('A')

elif conceito >= 7.5 < 9:
    print('{:.2f}'.format(conceito))
    print('B')

elif conceito >= 6 < 7.5:
    print('{:.2f}'.format(conceito))
    print('C')

elif conceito >= 5 < 6:
    print('{:.2f}'.format(conceito))
    print('D')

elif conceito < 5:
    print('{:.2f}'.format(conceito))
    print('E')
```

Uma fábrica vende discos voadores em pacotes fechados de três unidades. O prazo de entrega (em dias) é determinado pela combinação dos modelos escolhidos:

5 dias: Quando os três modelos são iguais (mesmo código).
15 dias: Quando apenas dois modelos são iguais e o terceiro é diferente.
30 dias: Quando todos os três modelos são diferentes entre si.

```python
def obter_prazo_entrega(disco1, disco2, disco3):
    prazo_entrega = 0

    if disco1 != disco2 and disco2 != disco3:
        prazo_entrega = 30

    if disco1 == disco2 or disco1 == disco3 or disco2 == disco3:
        prazo_entrega = 15

    if disco1 == disco2 and disco2 == disco3:
        prazo_entrega = 5
        
    return prazo_entrega

d1 = int(input())
d2 = int(input())
d3 = int(input())
prazo_entrega = obter_prazo_entrega(d1, d2, d3)
# print("Disco1 =", d1)
# print("Disco2 =", d2)
# print("Disco3 =", d3)
# print("Prazo de entrega =", prazo_entrega)
print(prazo_entrega)
```

Nesta atividade, você deve implementar a lógica para comparar duas datas cronologicamente. A função recebe 6 inteiros: d1, m1, a1 (Data 1) e d2, m2, a2 (Data 2).

O retorno deve ser:

-1: Se a Data 1 for a mais antiga.
0: Se as datas forem idênticas.
1: Se a Data 2 for a mais antiga.

```python
def comparar_datas(dia_1, mes_1, ano_1, dia_2, mes_2, ano_2):
    status = None
    if ano_2 > ano_1:
        status = -1
        
    if ano_2 == ano_1:
        if mes_2 < mes_1:
            status = 1
        if mes_2 > mes_1:
            status = -1
        if mes_2 == mes_1:
            if dia_2 > dia_1:
                status = -1
            if dia_2 > dia_1:
                status = 1
            elif dia_2 == dia_1:
                status = 0

    if ano_2 < ano_1:
        status = 1

    return status

comparar_datas(31,12,2021,1,1,2020)
```

A segurança dos passageiros depende da correta interpretação das regras. Reflita sobre a estrutura do seu código:

Para modelos entre 1901 e 2000, o que é mais importante: o motor ou a distância?
Na faixa de 2001 a 2020, como você verificaria apenas a distância, independente do motor instalado?
No caso específico de 2021, o que significa um critério que exige motor 200/201 E distância superior a 200? Ambas as condições precisam ser verdadeiras simultaneamente?
Como você lidaria com anos que não constam na tabela (ex: 2022)?

Ano de Produção	Critério Específico	Revisão?
1901 ≤ Ano ≤ 2000	Motor é 100 OU 101	SIM
2001 ≤ Ano ≤ 2020	Distância > 5000 anos-luz	SIM

Demais casos	Qualquer outra combinação	NAO
