Uma f√°brica vende discos voadores em pacotes fechados de tr√™s unidades. O prazo de entrega (em dias) √© determinado pela combina√ß√£o dos modelos escolhidos:

5 dias: Quando os tr√™s modelos s√£o iguais (mesmo c√≥digo).
15 dias: Quando apenas dois modelos s√£o iguais e o terceiro √© diferente.
30 dias: Quando todos os tr√™s modelos s√£o diferentes entre si.

```python
def obter_prazo_entrega(disco1, disco2, disco3):
    prazo_entrega = 0

    if disco1 != disco2 and disco2 != disco3:
        prazo_entrega = 30

    if disco1 == disco2 or disco1 == disco3 or disco2 == disco3:
        prazo_entrega = 15

    if disco1 == disco2 and disco2 == disco3:
        prazo_entrega = 5
        
    return prazo_entrega

d1 = int(input())
d2 = int(input())
d3 = int(input())
prazo_entrega = obter_prazo_entrega(d1, d2, d3)
print(prazo_entrega)
```

Nesta atividade, voc√™ deve implementar a l√≥gica para comparar duas datas cronologicamente. A fun√ß√£o recebe 6 inteiros: d1, m1, a1 (Data 1) e d2, m2, a2 (Data 2).

O retorno deve ser:

-1: Se a Data 1 for a mais antiga.
0: Se as datas forem id√™nticas.
1: Se a Data 2 for a mais antiga.

```python
def comparar_datas(dia_1, mes_1, ano_1, dia_2, mes_2, ano_2):
    status = None
    if ano_2 > ano_1:
        status = -1
        
    if ano_2 == ano_1:
        if mes_2 < mes_1:
            status = 1
        if mes_2 > mes_1:
            status = -1
        if mes_2 == mes_1:
            if dia_2 > dia_1:
                status = -1
            if dia_2 > dia_1:
                status = 1
            elif dia_2 == dia_1:
                status = 0

    if ano_2 < ano_1:
        status = 1

    return status
```
